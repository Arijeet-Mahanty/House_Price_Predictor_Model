### Importing necessary Python Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score
```

### Loading the dataset


```python
df = pd.read_csv("/content/sample_data/indian_house_prices.csv")
df.head()
```

### Encoding non-numeric data to numeric format

```python
le = LabelEncoder()
df["CityIndex"] = le.fit_transform(df["city"])
df["LocalityIndex"] = le.fit_transform(df["locality"])
df["property_typeIndex"] = le.fit_transform(df["property_type"])
df.head()
```

### Defining Features and Target

```python
X = df[['CityIndex','LocalityIndex','sqft_living','floors','balcony','property_typeIndex']]
Y = df['price']
```

### Splitting The Dataset

```python
X_train, X_test ,Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

### Train the linear model

```python
model = LinearRegression()
model.fit(X_train,Y_train)

Y_pred = model.predict(X_test)

print("R2 Score: ",r2_score(Y_test,Y_pred))
```

### Comparing the predicted data

```python
compare = pd.DataFrame({'Actual':Y_test,'Prediction':Y_pred.astype(int)})
compare.head()
```
